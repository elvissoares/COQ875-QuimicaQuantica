```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
```

## Átomo de Hélio

A função de onda tentativa do átomo de He é dada por 

$$ \psi(\boldsymbol{r}_1,\boldsymbol{r}_2) = \phi(\boldsymbol{r}_1)\phi(\boldsymbol{r}_2)$$

com as funções de Slater para o orbital 1s sendo 

$$\phi(\boldsymbol{r}) = \left(\frac{\zeta^3}{\pi}\right)^{1/2} e^{-\zeta r}$$

```python
def phi1s(r,zeta):
    norm = np.sqrt(zeta**3/np.pi)
    
    return norm*np.exp(-zeta*r)
```

```python
r = np.arange(0, 5, 0.01)  # Valores de r

plt.plot(r, phi1s(r, 2), label=r'$\zeta=2$')
plt.plot(r, phi1s(r, 17/16), label=r'$\zeta=27/16$')

plt.xlabel(r'$r$ (a.u.)')
plt.ylabel(r'$\phi_{1s}(r)$')

plt.xlim(0, 5)
plt.ylim(0, 1.6)
plt.legend(loc='upper right')
```

```python
plt.plot(r, 4*np.pi*phi1s(r, 2)**2*r**2, label=r'$\zeta=2$')
plt.plot(r, 4*np.pi*phi1s(r, 17/16)**2*r**2, label=r'$\zeta=27/16$')

plt.xlabel(r'$r$ (a.u.)')
plt.ylabel(r'$4\pi|\phi_{1s}(r)|^2 r^2$')

plt.xlim(0, 5)
plt.ylim(0, 1.2)
plt.legend(loc='upper right')
```
