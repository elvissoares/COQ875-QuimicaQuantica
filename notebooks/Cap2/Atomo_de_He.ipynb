```python
import numpy as np
import matplotlib.pyplot as plt
```

## Átomo de Hélio

A função de onda tentativa do átomo de He é dada por 

$$ \psi(\boldsymbol{r}_1,\boldsymbol{r}_2) = \phi(\boldsymbol{r}_1)\phi(\boldsymbol{r}_2)$$

com as funções de Slater para o orbital 1s sendo 

$$\phi(\boldsymbol{r}) = \left(\frac{\zeta^3}{\pi}\right)^{1/2} e^{-\zeta r}$$

tal que o valor esperado do Hamiltoniano é 

$$\langle \hat{H} \rangle = \zeta^2 - \frac{27}{8}\zeta$$

```python
def H(zeta):
    return zeta**2-(27/8)*zeta
```

```python
zetaarray = np.arange(0, 3.5, 0.01)  # Valores de zeta
Harray = H(zetaarray)  

zetastar = 27/16

plt.plot(zetaarray, Harray, label=r'$H(\zeta)$')
plt.axvline(zetastar, color='black', lw=0.5, ls='--')
plt.xlabel(r'$\zeta$')
plt.ylabel(r'$H(\zeta)$')
plt.title(r'Função de energia do átomo de He')
plt.legend()
```

```python
# função de onda 1s de Slater
def phi1s(r,zeta):
    norm = np.sqrt(zeta**3/np.pi)
    
    return norm*np.exp(-zeta*r)
```

```python
r = np.arange(0, 5, 0.01)  # Valores de r

plt.plot(r, phi1s(r, 2), label=r'$\zeta=2$')
plt.plot(r, phi1s(r, zetastar), label=fr'$\zeta={zetastar:.4f}$')

plt.xlabel(r'$r$ (a.u.)')
plt.ylabel(r'$\phi_{1s}(r)$')

plt.xlim(0, 5)
plt.ylim(0, 1.6)
plt.legend(loc='upper right')
```

```python
plt.plot(r, 4*np.pi*phi1s(r, 2)**2*r**2, label=r'$\zeta=2$')
plt.plot(r, 4*np.pi*phi1s(r, zetastar)**2*r**2, label=fr'$\zeta={zetastar:.4f}$')

plt.xlabel(r'$r$ (a.u.)')
plt.ylabel(r'$4\pi|\phi_{1s}(r)|^2 r^2$')

plt.xlim(0, 5)
plt.ylim(0, 1.2)
plt.legend(loc='upper right')
```
