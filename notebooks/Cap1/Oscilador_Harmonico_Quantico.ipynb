```python
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
```

# Função de onda do oscilador harmônico quântico

$\psi_n(x) = \frac{1}{(2^n n!)^{1/2}}\left(\frac{\alpha}{\pi}\right)^{1/4} H_n(\alpha^{1/2} x) e^{-\alpha x^2/2}$ com $\alpha = \frac{m \omega}{\hbar}$

$H_n(x)$: Polinômios de Hermite (Scipy)

```python
def H(n,x):
    """Calculates the Hermite polynomial of degree n at point x."""
    return special.hermite(n)(x)
```

## Polinômios de Hermite

$$H_n(x) = (-1)^n e^{x^2} \left( \frac{d}{d x}\right)^n e^{-x^2}$$

```python
x = np.arange(-2,2,0.001)

plt.plot(x, H(0,x))
plt.plot(x, H(1,x))
plt.plot(x, H(2,x))
plt.plot(x, H(3,x))
plt.title("Polinômios de Hermite")
plt.xlabel("x")
plt.ylabel(r"$H_n(x)$")
plt.legend(['H0','H1', 'H2', 'H3'])
```

```python
def psi_n(n, x, alpha):
    """Calculates the wave function of the quantum harmonic oscillator."""
    norm = (1 / np.sqrt(2**n * special.factorial(n))) * (alpha / np.pi)**0.25
    return norm * H(n, alpha**0.5 * x) * np.exp(-alpha * x**2 / 2)
```

```python
x = np.arange(-5,5,0.001)

plt.figure(figsize=(4, 8))
plt.plot(x, psi_n(0, x, 1))
plt.plot(x, psi_n(1, x, 1)+1)
plt.plot(x, psi_n(2, x, 1)+2)
plt.plot(x, psi_n(3, x, 1)+3)
plt.title("Funções de onda do oscilador harmônico quântico")
plt.xlabel("x")
plt.ylabel(r"$\psi_n(x)$")
plt.legend([r"$\psi_0$", r"$\psi_1$", r"$\psi_2$", r"$\psi_3$"])      
```

```python
plt.figure(figsize=(4, 8))
plt.plot(x, psi_n(0, x, 1)**2)
plt.plot(x, psi_n(1, x, 1)**2+1)
plt.plot(x, psi_n(2, x, 1)**2+2)
plt.plot(x, psi_n(3, x, 1)**2+3)
plt.title("Densidade de probabilidade do oscilador harmônico quântico")
plt.xlabel("x")
plt.ylabel(r"$|\psi_n(x)|^2$")
plt.legend([r"$\psi_0$", r"$\psi_1$", r"$\psi_2$", r"$\psi_3$"])     
```
